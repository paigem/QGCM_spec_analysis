```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

# Load and plot spectral energy terms

```python
# Load the terms (for year 013)
TKE1_013 = np.load('TKE1_output013_clean_correctsign.npy')
TKE2_013 = np.load('TKE2_output013_clean_correctsign.npy')
TKE3_013 = np.load('TKE3_output013_clean_correctsign.npy')

TPE1_013 = np.load('TPE1_output013_clean_correctsign.npy')
TPE2_013 = np.load('TPE2_output013_clean_correctsign.npy')

Twindstress_013 = np.load('Twindstress_output013_clean_correctsign.npy')
Tbuoyancy_013 = np.load('Tbuoyancy_output013_clean_correctsign.npy')
TbottomDrag_013 = np.load('TbottomDrag_output013_clean_correctsign.npy')
```

## Plot all terms

```python
ktiso = (2*np.pi/(365*24*60*60))*np.arange(0,np.floor(365.0/2)+1)

plt.figure(figsize = (12,8))

#plt.semilogx(TKE1_013*ktiso*1e9,color='b',label='KE1')
#plt.semilogx(TKE2_013*ktiso*1e9,color='b',linestyle='dashed',label='KE2')
#plt.semilogx(TKE3_013*ktiso*1e9,color='b',linestyle='dotted',label='KE3')

plt.semilogx(TPE1_013*ktiso*1e9,color='m',label='PE1')
plt.semilogx(TPE2_013*ktiso*1e9,color='m',linestyle='dashed',label='PE2')

plt.semilogx(Twindstress_013*ktiso*1e9,color='c',label='windstress')
plt.semilogx(Tbuoyancy_013*ktiso*1e9,color='g',label='buoyancy')
plt.semilogx(TbottomDrag_013*ktiso*1e9,color='y',label='bottom drag')

plt.semilogx((-TPE1_013-TPE2_013-Twindstress_013-Tbuoyancy_013-TbottomDrag_013)*ktiso*1e9,'k',label='residual')

plt.axhline(0,color='k',linestyle='dotted')

plt.title('Transfer budget (no KE)')
plt.xlabel('Frequency (rad/day)')
plt.ylabel('Spectral transfer (nW/kg)')
plt.legend()
```

## Plot just KE terms

```python
ktiso = (2*np.pi/(365*24*60*60))*np.arange(0,np.floor(365.0/2)+1)

plt.figure(figsize = (12,8))

plt.semilogx(TKE1_013*ktiso*1e9,color='b',label='KE1')
plt.semilogx(TKE2_013*ktiso*1e9,color='b',linestyle='dashed',label='KE2')
plt.semilogx(TKE3_013*ktiso*1e9,color='b',linestyle='dotted',label='KE3')

plt.semilogx((-TPE1_013-TPE2_013-Twindstress_013-Tbuoyancy_013-TbottomDrag_013)*ktiso*1e9,'k',label='residual')
plt.semilogx((TKE1_013 + TKE2_013 + TKE3_013)*ktiso*1e9,'r',label='KE sum')

plt.axhline(0,color='k',linestyle='dotted')

plt.title('Transfer budget (only KE)')
plt.xlabel('Frequency (rad/day)')
plt.ylabel('Spectral transfer (nW/kg)')
plt.legend()
```
