```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

# Load and plot spectral energy terms

```python
# Load the terms (for years 013 - 019)
TKE1 = np.load('TKE1_output013_019.npy')
TKE2 = np.load('TKE2_output013_019.npy')
TKE3 = np.load('TKE3_output013_019.npy')

TPE1 = np.load('TPE1_output013_019.npy')
TPE2 = np.load('TPE2_output013_019.npy')

Twindstress = np.load('Twindstress_output013_019.npy')
Tbuoyancy = np.load('Tbuoyancy_output013_019.npy')
TbottomDrag = np.load('TbottomDrag_output013_019.npy')

#---------------------
# Load the terms (for years 013 - 019) with new BC
TKE1_newBC = np.load('TKE1_output013_019_newBC2_correctderiv.npy')
TKE2_newBC = np.load('TKE2_output013_019_newBC2_correctderiv.npy')
TKE3_newBC = np.load('TKE3_output013_019_newBC2_correctderiv.npy')

TPE1_newBC = np.load('TPE1_output013_019_newBC.npy')
TPE2_newBC = np.load('TPE2_output013_019_newBC.npy')

TbottomDrag_newBC = np.load('TbottomDrag_output013_019_newBC.npy')


#---------------------
# Load the terms (for year 013)
TKE1_013 = np.load('TKE1_output013_newBC.npy')
TKE2_013 = np.load('TKE2_output013_newBC.npy')
TKE3_013 = np.load('TKE3_output013_newBC.npy')

TPE1_013 = np.load('TPE1_output013_newBC.npy')
TPE2_013 = np.load('TPE2_output013_newBC.npy')

Twindstress_013 = np.load('Twindstress_output013.npy')
Tbuoyancy_013 = np.load('Tbuoyancy_output013.npy')
TbottomDrag_013 = np.load('TbottomDrag_output013_newBC.npy')
```

```python
ktiso = (2*np.pi/(365*24*60*60))*np.arange(0,np.floor(7*365.0/2)+1)

plt.figure(figsize = (12,8))

plt.semilogx(-TPE1*ktiso*1e9,color='m',label='PE1')
plt.semilogx(-TPE2*ktiso*1e9,color='m',linestyle='dashed',label='PE2')

plt.semilogx(Twindstress*ktiso*1e9,color='c',label='windstress')
plt.semilogx(Tbuoyancy*ktiso*1e9,color='g',label='buoyancy')
plt.semilogx(-TbottomDrag*ktiso*1e9,color='y',label='bottom drag')

plt.axhline(0,color='k',linestyle='dotted')

plt.title('Transfer budget (no KE)')
plt.xlabel('Frequency (rad/day)')
plt.ylabel('Spectral transfer (nW/kg)')
plt.legend()
```
